### Prepare Phase

**FitBit Fitness Tracker Data**: This Kaggle data set
contains personal fitness tracker from thirty fitbit users. Thirty eligible Fitbit users consented to the submission of personal tracker data, including minute-level output for physical activity, heart rate, and sleep monitoring. It includes information about daily activity, steps, and heart rate that can be used to explore usersâ€™ habits.

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
DATA_PARENT_FOLDER = "Fitabase_Data_4_12_2016_5_12_2016/"
```

```python
data_dict = {}
```

```python
for filename in sorted(os.listdir(DATA_PARENT_FOLDER)):
    name = filename[:-4]
    print(name)
    data_dict[name] = pd.read_csv(DATA_PARENT_FOLDER+filename)
    display(data_dict[name])
    print("-"*100)
```

#### Data description
Data is mainly collected between 12th of April - 12th of May 2016.
* dailyActivity_merged - for each user ID number of steps, distance (active, fairly active, lightly active, sedentary), activity minutes (active, fairly active, lightly active, sedentary), and total number of calories per day.
* dailyCalories_merged - for each user Id total number of calories per day
* dailyIntensities_merged - for each user Id distance (active, fairly active, lightly active, sedentary), activity minutes (active, fairly active, lightly active, sedentary) per day
* dailySteps_merged - for each Id number of steps per day
* heartrate_seconds_merged - heart rate of user measured roughly every 5 seconds
* hourlyCalories_merged - number of calories burnt per user on each hour
* hourlyIntensities_merged - intensity of user total and average each hour
* hourlySteps_merged - number of steps of user on each hour
* minuteCaloriesNarrow_merged - calories of user per minute
* minuteCaloriesWide_merged - the same as minuteCaloriesNarrow_merged but each minute of an hour in one row
* minuteIntensitiesNarrow_merged - intensity of user per minute
* minuteIntensitiesWide_merged - the same as minuteIntensitiesNarrow_merged but each minute of an hour in one row
* minuteMETsNarrow_merged - MET (Metabolic Equivalent, shows intensity of an activity) for each user measured every minute
* minuteSleep_merged - minutes of sleep of each user
* minuteStepsNarrow_merged - number of steps of each user per minute
* minuteStepsWide_merged - the same as minuteStepsNarrow_merged but each minute of an hour in one row
* sleepDay_merged - for each user number of minutes slept a day and total time in bed
* weightLogInfo_merged - measurements of some users: weight in kg and pounds, bmi (body mass index), fat

### Analyze phase

Let's see how accurate the tracker calculates the distance.

```python
tracker_distance = data_dict["dailyActivity_merged"]["TrackerDistance"]
total_distance = data_dict["dailyActivity_merged"]["TotalDistance"]
```

```python
diff_distance_tracked_percentage = (total_distance - tracker_distance)/total_distance*100
```

```python
np.mean(diff_distance_tracked_percentage)
```

```python
np.std(diff_distance_tracked_percentage)
```

**Discussion**

We show that the mean percentage error of tracker in distance is 0.14% Â± 1.2%.

```python
total_steps = data_dict["dailyActivity_merged"]["TotalSteps"]
total_calories = data_dict["dailyActivity_merged"]["Calories"]
```

```python
np.corrcoef(total_steps, total_calories)[0, 1]
```

```python
ax = sns.regplot(x="TotalSteps", 
                y="Calories", 
                data=data_dict["dailyActivity_merged"], 
                scatter_kws={'alpha':0.2}, line_kws={"color": "red"})
plt.xlabel("Total number of steps")
plt.ylabel("Calories")
plt.title("Daily Total Number of Steps vs Calories")
plt.show()
```

**Discussion**

We see that there is positive dependence (red line) between the number of steps done per day and the amount of the calories burnt.

```python
data_dict["sleepDay_merged"]["SleepDay"] = data_dict["sleepDay_merged"]["SleepDay"].str[:-12]
```

```python
data_dict["sleepDay_merged"].rename(columns={"SleepDay": "ActivityDate"}, inplace=True)
```

```python
# it seems there are some duplicates in the sleepDay_merged table
data_dict["sleepDay_merged"][data_dict["sleepDay_merged"].duplicated(keep=False)]
```

```python
#remove duplicates
data_dict["sleepDay_merged"].drop_duplicates(["Id", "ActivityDate"], inplace=True)
```

```python
# we add information about the sleep to the daily activity table
data_dict["dailyActivity_merged"] = data_dict["dailyActivity_merged"].merge(data_dict["sleepDay_merged"], how='left', on=['Id', 'ActivityDate'])
```

```python
data_dict["dailyActivity_merged"]
```

```python
data_dict["dailyActivity_merged"]["TimeInBedNotSleeping"] = \
(data_dict["dailyActivity_merged"]["TotalTimeInBed"] - data_dict["dailyActivity_merged"]["TotalMinutesAsleep"])
```

```python
data_dict["dailyActivity_merged"][data_dict["dailyActivity_merged"]["TotalMinutesAsleep"].notna()]
```

```python
data_dict["dailyActivity_merged"]["TotalMinutes"] = data_dict["dailyActivity_merged"]["VeryActiveMinutes"] + \
data_dict["dailyActivity_merged"]["FairlyActiveMinutes"] + data_dict["dailyActivity_merged"]["LightlyActiveMinutes"] + \
data_dict["dailyActivity_merged"]["SedentaryMinutes"]
```

```python
data_dict["dailyActivity_merged"][data_dict["dailyActivity_merged"]["TotalMinutesAsleep"].notna()]
```

```python
ax = sns.scatterplot(x="TotalSteps", 
                y="TimeInBedNotSleeping", 
                data=data_dict["dailyActivity_merged"][data_dict["dailyActivity_merged"]["TotalMinutesAsleep"].notna()], 
                alpha = 1.0)
plt.xlabel("Total number of steps")
plt.ylabel("Time in Bed Not Sleeping")
plt.title("Daily Total Number of Steps vs Quality of Sleep")
plt.show()
```

**Discussion**
We see that people with higher number of steps per day spend less time on bed just laying, i.e. the quality of sleep increases.
